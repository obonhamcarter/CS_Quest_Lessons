## 🎯 The Computer Science Major

**For students who want to become professional software developers, researchers, or tech leaders.**

### What You'll Learn

**🧱 Foundations**
- Programming (Python, Java, C++)
- Data Structures & Algorithms
- Computer Architecture
- Operating Systems

**🤖 Advanced Topics**
- Artificial Intelligence
- Machine Learning
- Database Systems
- Software Engineering

**🎨 Specializations**
- Web Development
- Mobile Apps
- Computer Graphics
- Cybersecurity

**🔬 Research**
- Independent Study
- Senior Thesis
- Faculty Collaboration
- Conference Presentations

### Sample Course Sequence

**First Year:**
- CMPSC 100: Computational Expression
- CMPSC 101: Data Abstraction
- Math courses (Discrete Structures, Calculus)

**Second Year:**
- Data Structures
- Computer Organization
- Web Development
- Choose electives aligned with interests

**Third Year:**
- Algorithm Analysis
- Software Engineering
- Database Systems
- Specialized electives (AI, Graphics, etc.)

**Fourth Year:**
- Advanced electives
- Senior Thesis/Capstone Project
- Internship opportunities

### Career Outcomes

**CS Majors typically become:**
- **Software Engineers** at companies like Google, Microsoft, Amazon, Meta
- **Data Scientists** at healthcare, finance, and tech companies
- **AI/ML Engineers** developing intelligent systems
- **Cybersecurity Specialists** protecting organizations
- **Graduate Students** pursuing MS or PhD in CS
- **Tech Entrepreneurs** starting their own companies
- **Research Scientists** at universities or labs

**💼 Average Starting Salary for CS Majors: $75,000 - $120,000**

*Note: Varies by location, company, and specialization*

## 💡 The Computer Science Minor

**Perfect for students who want to combine CS skills with another major to create unique career opportunities!**

### Why Choose a CS Minor?

**CS skills are valuable in EVERY field:**
- 🧬 **Biology + CS** = Bioinformatics, Genomics Research, Medical Software
- 💼 **Business + CS** = Tech Entrepreneurship, Data Analytics, Product Management
- 🎨 **Art + CS** = Game Design, Digital Art, UX/UI Design, Animation
- 🌍 **Environmental Science + CS** = Climate Modeling, Conservation Tech
- 📚 **Education + CS** = EdTech Development, Digital Learning Platforms
- ⚖️ **Political Science + CS** = Policy Analysis, Civic Tech, Data Journalism
- 🎵 **Music + CS** = Audio Software, Music Production Apps, Sound Design

### What You'll Learn

**Core Requirements (Foundation):**
- Programming fundamentals (Python, Java)
- Data Structures
- One or two intermediate CS courses

**Electives (Choose Based on Your Interests):**
- Web Development → Build websites and web apps
- Database Systems → Manage and analyze data
- AI/Machine Learning → Create intelligent systems
- Mobile App Development → Build iOS/Android apps
- Computer Graphics → Create visual applications
- Data Science → Analyze and visualize data

### Sample Combinations & Careers

**🧬 Biology + CS Minor**
- Career: Bioinformatics Scientist
- Use CS to: Analyze genomic data, model biological systems, develop medical software
- Real Companies: 23andMe, Illumina, Moderna

**💼 Business + CS Minor**
- Career: Product Manager / Tech Entrepreneur
- Use CS to: Understand technical constraints, communicate with engineers, analyze business data
- Real Companies: Any tech company, startups, consulting

**🎨 Art + CS Minor**
- Career: Game Designer / Digital Artist
- Use CS to: Create interactive experiences, build game mechanics, develop art tools
- Real Companies: Epic Games, Pixar, Adobe, game studios

**📊 Psychology + CS Minor**
- Career: UX Researcher / HCI Specialist
- Use CS to: Design user interfaces, conduct usability research, build accessible software
- Real Companies: Google UX, Apple, Meta, design agencies

## 🤔 How Do I Choose?

### Choose the CS MAJOR if you:
- ✅ Want to become a professional software developer
- ✅ Are fascinated by how computers work
- ✅ Want to pursue graduate school in CS
- ✅ Love problem-solving and building things with code
- ✅ Want maximum flexibility in tech careers

### Choose the CS MINOR if you:
- ✅ Have a primary passion in another field
- ✅ Want to enhance your major with tech skills
- ✅ Want to be competitive in any career (data skills are universal!)
- ✅ Are interested in interdisciplinary work
- ✅ Want to understand technology without becoming a full-time developer

### 💡 Can't Decide?

**Start with introductory CS courses!** You can always:
- Begin as a CS minor and upgrade to a major
- Take a few CS courses just to explore
- Double major (yes, it's possible!)
- Talk to current students and faculty

**The important thing is to START LEARNING!**

## 📚 Course Load & Time Commitment

### CS Major
- **~12-14 CS courses** + supporting math courses
- **3-4 CS courses per year** (manageable with any major/minor combo)
- **Senior Thesis/Project** (meaningful capstone experience)

### CS Minor
- **~6-7 CS courses**
- **1-2 CS courses per year** (easy to fit with any major)
- **No thesis required** (but research opportunities available)

**Both options are designed to fit comfortably with:**
- Other majors/minors
- Study abroad
- Athletics
- Research opportunities
- Internships
- Campus involvement
