## 🎨 Demo 1: Draw with Code

See how code can create art! This program draws a colorful spiral.

**Try changing:**
- The colors to your favorites (try: 'purple', 'green', 'orange', 'pink', 'gold')
- The number `59` to `90` or `45` - what happens?
- Change `i * 4` to `i * 6` to make it bigger!

````python
# Draw a spiral with Python!
# Try changing the colors below:
colors = ['red', 'yellow', 'blue']  # Change these!

import turtle
turtle.speed('fastest')
turtle.bgcolor('black')

for i in range(50):
    turtle.color(colors[i % len(colors)])
    turtle.forward(i * 4)
    turtle.right(59)  # Try changing this number!

turtle.hideturtle()
print("🎨 Art created with code!")
````

## 🎉 Demo 4: Birthday Calculator

Calculate how many days until your next birthday!

````python
# Birthday Calculator
from datetime import datetime, timedelta

print("🎂 BIRTHDAY COUNTDOWN")
print("=" * 30)

# Get today's date
today = datetime.now()
current_year = today.year

# Enter your birthday (change these numbers!)
birth_month = 7   # <- Change this (1-12)
birth_day = 15    # <- Change this (1-31)

# Calculate next birthday
try:
    next_birthday = datetime(current_year, birth_month, birth_day)
    
    # If birthday already passed this year, use next year
    if next_birthday < today:
        next_birthday = datetime(current_year + 1, birth_month, birth_day)
    
    days_until = (next_birthday - today).days
    
    print(f"📅 Today: {today.strftime('%B %d, %Y')}")
    print(f"🎉 Next Birthday: {next_birthday.strftime('%B %d, %Y')}")
    print(f"⏰ Days until birthday: {days_until}")
    
    if days_until == 0:
        print("🎊 HAPPY BIRTHDAY! 🎊")
    elif days_until < 7:
        print("🎈 Your birthday is this week!")
    elif days_until < 30:
        print("🎁 Your birthday is this month!")
    
except ValueError:
    print("❌ Invalid date! Check month (1-12) and day (1-31)")
````

## 🧮 Demo 5: Password Strength Checker

Is your password strong enough? Let's check!

````python
# Password Strength Checker
def check_password_strength(password):
    """Check how strong a password is"""
    score = 0
    feedback = []
    
    # Check length
    if len(password) >= 8:
        score += 2
        feedback.append("✅ Good length (8+ characters)")
    else:
        feedback.append("❌ Too short (needs 8+ characters)")
    
    # Check for numbers
    if any(char.isdigit() for char in password):
        score += 1
        feedback.append("✅ Contains numbers")
    else:
        feedback.append("⚠️  Add some numbers")
    
    # Check for uppercase
    if any(char.isupper() for char in password):
        score += 1
        feedback.append("✅ Contains uppercase letters")
    else:
        feedback.append("⚠️  Add uppercase letters")
    
    # Check for lowercase
    if any(char.islower() for char in password):
        score += 1
        feedback.append("✅ Contains lowercase letters")
    
    # Check for special characters
    special_chars = "!@#$%^&*()_+-=[]{}|;:',.<>?"
    if any(char in special_chars for char in password):
        score += 1
        feedback.append("✅ Contains special characters")
    else:
        feedback.append("⚠️  Add special characters (!@#$)")
    
    return score, feedback

# Test some passwords
print("🔐 PASSWORD STRENGTH CHECKER")
print("=" * 40)

test_passwords = [
    "password",
    "Password123",
    "MyP@ssw0rd!",
]

for pwd in test_passwords:
    score, feedback = check_password_strength(pwd)
    
    print(f"\n🔑 Testing: '{pwd}'")
    print(f"💪 Strength: {score}/6")
    
    if score <= 2:
        print("🔴 Status: WEAK")
    elif score <= 4:
        print("🟡 Status: MODERATE")
    else:
        print("🟢 Status: STRONG")
    
    print("\nFeedback:")
    for item in feedback:
        print(f"  {item}")

print("\n" + "=" * 40)
print("🛡️  This is how real apps check password strength!")
````
