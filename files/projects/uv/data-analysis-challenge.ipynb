## Step 1: Import Libraries

First, let's import the tools we'll need for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("✅ Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
```

## Step 2: Create Sample Survey Data

Let's create a dataset representing a survey of 25 students. Each student shared:
- Their name
- Favorite tech topic (AI, Web Dev, Game Dev, Data Science, or Cybersecurity)  
- Hours they study per week
- Experience level (Beginner, Intermediate, or Advanced)

```python
# Create the survey data
data = {
    'name': ['Alex', 'Blake', 'Casey', 'Drew', 'Ember', 'Finley', 'Gray', 'Harper',
             'Jordan', 'Kelly', 'Logan', 'Morgan', 'Noel', 'Parker', 'Quinn',
             'Riley', 'Sage', 'Taylor', 'Uma', 'Val', 'West', 'Xen', 'Yara', 'Zoe', 'Avery'],
    'favorite_topic': ['AI', 'Web Dev', 'Game Dev', 'Data Science', 'AI',
                       'Cybersecurity', 'Web Dev', 'AI', 'Game Dev', 'Data Science',
                       'AI', 'Web Dev', 'Game Dev', 'Cybersecurity', 'Data Science',
                       'AI', 'Game Dev', 'Web Dev', 'Cybersecurity', 'Data Science',
                       'AI', 'Game Dev', 'Web Dev', 'Data Science', 'AI'],
    'hours_per_week': [10, 5, 15, 8, 12, 6, 7, 14, 20, 9, 11, 4, 18, 7, 10,
                       13, 16, 8, 5, 12, 9, 22, 6, 11, 15],
    'experience_level': ['Intermediate', 'Beginner', 'Advanced', 'Intermediate', 'Intermediate',
                         'Beginner', 'Beginner', 'Intermediate', 'Advanced', 'Intermediate',
                         'Intermediate', 'Beginner', 'Advanced', 'Beginner', 'Intermediate',
                         'Advanced', 'Advanced', 'Intermediate', 'Beginner', 'Intermediate',
                         'Intermediate', 'Advanced', 'Beginner', 'Intermediate', 'Advanced']
}

# Create DataFrame
df = pd.DataFrame(data)
print("✅ Data created successfully!")
print(f"Total students surveyed: {len(df)}")
```

## Step 3: Explore the Data

Let's look at our data and understand what we're working with!

```python
# View the first few rows
print("📋 First 5 students in our survey:")
df.head()
```

```python
# Count how many students chose each topic
print("📊 Topic Popularity:")
topic_counts = df['favorite_topic'].value_counts()
print(topic_counts)
print(f"\nMost popular topic: {topic_counts.idxmax()} with {topic_counts.max()} students")
```

```python
# Calculate statistics about hours studied
print("⏰ Study Hours Statistics:")
print(f"Average hours per week: {df['hours_per_week'].mean():.1f}")
print(f"Minimum hours: {df['hours_per_week'].min()}")
print(f"Maximum hours: {df['hours_per_week'].max()}")
print(f"Median hours: {df['hours_per_week'].median()}")
```

## Step 4: Create Visualizations

Now let's make some charts to visualize our findings!

```python
# Create a bar chart of topic popularity
plt.figure(figsize=(10, 6))
topic_counts.plot(kind='bar', color='skyblue', edgecolor='navy')
plt.title('Student Favorite Tech Topics', fontsize=16, fontweight='bold')
plt.xlabel('Tech Topic', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"🎯 {topic_counts.idxmax()} is the winner!")
```

```python
# Create a histogram of study hours
plt.figure(figsize=(10, 6))
plt.hist(df['hours_per_week'], bins=10, color='lightcoral', edgecolor='darkred', alpha=0.7)
plt.title('Distribution of Weekly Study Hours', fontsize=16, fontweight='bold')
plt.xlabel('Hours per Week', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.axvline(df['hours_per_week'].mean(), color='red', linestyle='--', linewidth=2, label=f'Average: {df["hours_per_week"].mean():.1f}h')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
