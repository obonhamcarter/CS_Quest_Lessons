## Part 1: Grid-Based Coordinates

Many classic games use a grid system. Think of it like a spreadsheet - every position has an (x, y) coordinate!

```python
# Define a player position on a 10x10 grid
player_x = 5
player_y = 5
grid_width = 10
grid_height = 10

print(f"🎮 Player is at position ({player_x}, {player_y})")
print(f"Grid size: {grid_width} x {grid_height}")

# Visualize the grid (simple text representation)
for y in range(grid_height):
    row = ""
    for x in range(grid_width):
        if x == player_x and y == player_y:
            row += "P "  # P for player
        else:
            row += ". "  # . for empty space
    print(row)

print("\nP = Player position")
```

## Part 2: Movement and Boundary Checking

Let's create functions to move the player and make sure they stay within the grid!

```python
def move_player(x, y, direction, grid_width, grid_height):
    """Move player in a direction, checking boundaries"""
    new_x, new_y = x, y
    
    if direction == "up":
        new_y -= 1
    elif direction == "down":
        new_y += 1
    elif direction == "left":
        new_x -= 1
    elif direction == "right":
        new_x += 1
    
    # Check boundaries
    if 0 <= new_x < grid_width and 0 <= new_y < grid_height:
        return new_x, new_y
    else:
        print("⚠️ Can't move there - out of bounds!")
        return x, y

# Test the movement
player_x, player_y = 5, 5
print(f"Starting position: ({player_x}, {player_y})")

player_x, player_y = move_player(player_x, player_y, "up", 10, 10)
print(f"After moving up: ({player_x}, {player_y})")

player_x, player_y = move_player(player_x, player_y, "right", 10, 10)
print(f"After moving right: ({player_x}, {player_y})")
```

## Part 3: Collision Detection

Collision detection is crucial for games - it tells us when two objects overlap!

```python
def check_collision(x1, y1, x2, y2):
    """Check if two objects at the same position collide"""
    return x1 == x2 and y1 == y2

def check_rect_collision(x1, y1, width1, height1, x2, y2, width2, height2):
    """Check if two rectangles overlap (for larger objects)"""
    return (x1 < x2 + width2 and
            x1 + width1 > x2 and
            y1 < y2 + height2 and
            y1 + height1 > y2)

# Test point collision
player_pos = (5, 5)
food_pos = (5, 5)
enemy_pos = (3, 7)

print(f"Player at {player_pos}")
print(f"Food at {food_pos}: Collision? {check_collision(*player_pos, *food_pos)}")
print(f"Enemy at {enemy_pos}: Collision? {check_collision(*player_pos, *enemy_pos)}")

# Test rectangle collision
print("\nRectangle Collision Test:")
player_rect = (5, 5, 2, 2)  # x, y, width, height
obstacle_rect = (6, 6, 2, 2)
print(f"Player overlaps obstacle? {check_rect_collision(*player_rect, *obstacle_rect)}")
```

## Part 4: Game State Management

Games need to track lots of information - player position, score, lives, etc. Let's use dictionaries!

```python
# Create a game state dictionary
game_state = {
    'player': {'x': 5, 'y': 5, 'lives': 3},
    'score': 0,
    'level': 1,
    'is_running': True,
    'enemies': [
        {'x': 2, 'y': 3},
        {'x': 8, 'y': 7}
    ],
    'food': {'x': 7, 'y': 4}
}

def print_game_state(state):
    """Display the current game state"""
    print("=== GAME STATE ===")
    print(f"Player: ({state['player']['x']}, {state['player']['y']})")
    print(f"Lives: {'❤️ ' * state['player']['lives']}")
    print(f"Score: {state['score']}")
    print(f"Level: {state['level']}")
    print(f"Enemies: {len(state['enemies'])}")
    print(f"Food at: ({state['food']['x']}, {state['food']['y']})")
    print("==================\n")

print_game_state(game_state)

# Update game state
game_state['score'] += 10
game_state['player']['x'] += 1
print("After moving and scoring:")
print_game_state(game_state)
```

```python
# Challenge 1: Snake Movement
# YOUR CODE HERE

def move_snake(snake, direction):
    """Move snake by adding new head and removing tail"""
    head_x, head_y = snake[0]
    
    if direction == "up":
        new_head = (head_x, head_y - 1)
    elif direction == "down":
        new_head = (head_x, head_y + 1)
    elif direction == "left":
        new_head = (head_x - 1, head_y)
    elif direction == "right":
        new_head = (head_x + 1, head_y)
    
    # Add new head at the beginning
    snake.insert(0, new_head)
    # Remove tail
    snake.pop()
    return snake

# Test it!
snake = [(5, 5), (5, 6), (5, 7)]
print(f"Snake before: {snake}")
snake = move_snake(snake, "right")
print(f"Snake after moving right: {snake}")
```
