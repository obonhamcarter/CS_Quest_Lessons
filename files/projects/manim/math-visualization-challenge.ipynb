## Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

print("✅ Libraries imported!")
print(f"NumPy version: {np.__version__}")
```

## Step 2: Plot Basic Mathematical Functions

Let's visualize some classic mathematical functions!

```python
# Create x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate y values for different functions
y_sin = np.sin(x)
y_cos = np.cos(x)
y_parabola = x**2 / 10

# Plot all functions
plt.figure(figsize=(12, 6))

plt.plot(x, y_sin, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', color='red', linewidth=2)
plt.plot(x, y_parabola, label='x²/10', color='green', linewidth=2)

plt.title('Mathematical Functions', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

print("📈 Functions plotted!")
```

## Step 3: Create Geometric Shapes

Let's create circles, polygons, and other shapes using parametric equations!

```python
def create_circle(radius, center=(0, 0), num_points=100):
    """Create a circle using parametric equations"""
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = center[0] + radius * np.cos(theta)
    y = center[1] + radius * np.sin(theta)
    return x, y

def create_polygon(n_sides, radius, center=(0, 0)):
    """Create a regular polygon"""
    angles = np.linspace(0, 2 * np.pi, n_sides + 1)
    x = center[0] + radius * np.cos(angles)
    y = center[1] + radius * np.sin(angles)
    return x, y

# Create shapes
plt.figure(figsize=(12, 5))

# Subplot 1: Circles
plt.subplot(1, 2, 1)
for radius in [1, 2, 3]:
    x, y = create_circle(radius)
    plt.plot(x, y, linewidth=2, label=f'r={radius}')
plt.title('Concentric Circles', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.legend()

# Subplot 2: Polygons
plt.subplot(1, 2, 2)
for sides in [3, 5, 6, 8]:
    x, y = create_polygon(sides, radius=2)
    plt.plot(x, y, linewidth=2, marker='o', label=f'{sides} sides')
plt.title('Regular Polygons', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

print("✨ Geometric shapes created!")
```

## Step 4: Transformations (Translation, Rotation, Scaling)

Transformations are the key to animation! Let's see how to move, rotate, and scale shapes.

```python
def translate(x, y, dx, dy):
    """Translate (move) points by dx, dy"""
    return x + dx, y + dy

def scale(x, y, factor):
    """Scale points by a factor"""
    return x * factor, y * factor

def rotate(x, y, angle):
    """Rotate points by angle (in radians)"""
    cos_a = np.cos(angle)
    sin_a = np.sin(angle)
    x_new = x * cos_a - y * sin_a
    y_new = x * sin_a + y * cos_a
    return x_new, y_new

# Demonstrate transformations
plt.figure(figsize=(15, 5))

# Original square
square_x = np.array([0, 1, 1, 0, 0])
square_y = np.array([0, 0, 1, 1, 0])

# Translation
plt.subplot(1, 3, 1)
plt.plot(square_x, square_y, 'b-', linewidth=2, label='Original')
trans_x, trans_y = translate(square_x, square_y, 2, 1)
plt.plot(trans_x, trans_y, 'r--', linewidth=2, label='Translated (2,1)')
plt.title('Translation', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()

# Scaling
plt.subplot(1, 3, 2)
plt.plot(square_x, square_y, 'b-', linewidth=2, label='Original')
scaled_x, scaled_y = scale(square_x, square_y, 1.5)
plt.plot(scaled_x, scaled_y, 'g--', linewidth=2, label='Scaled x1.5')
plt.title('Scaling', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()

# Rotation
plt.subplot(1, 3, 3)
plt.plot(square_x, square_y, 'b-', linewidth=2, label='Original')
rotated_x, rotated_y = rotate(square_x, square_y, np.pi / 4)  # 45 degrees
plt.plot(rotated_x, rotated_y, 'm--', linewidth=2, label='Rotated 45°')
plt.title('Rotation', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()

plt.tight_layout()
plt.show()

print("🔄 Transformations demonstrated!")
```

```python
# Challenge 2: Pythagorean Theorem Visualization
# YOUR CODE HERE

plt.figure(figsize=(8, 8))

# Draw right triangle
a, b = 3, 4
c = np.sqrt(a**2 + b**2)

# Triangle vertices
triangle_x = [0, a, 0, 0]
triangle_y = [0, 0, b, 0]
plt.plot(triangle_x, triangle_y, 'k-', linewidth=3, label='Triangle')

# Square on side a
square_a_x = [a, a+b, a+b, a, a]
square_a_y = [0, 0, a, a, 0]
plt.fill(square_a_x, square_a_y, alpha=0.3, color='blue', label=f'a² = {a**2}')

# Square on side b
square_b_x = [0, 0, b, b, 0]
square_b_y = [b, b+a, b+a, b, b]
plt.fill(square_b_x, square_b_y, alpha=0.3, color='red', label=f'b² = {b**2}')

# Square on hypotenuse c (rotated)
angle = np.arctan(b/a)
square_c_x = [a, a + c*np.cos(angle), a + c*np.cos(angle) - c*np.sin(angle), a - c*np.sin(angle), a]
square_c_y = [0, c*np.sin(angle), c*np.sin(angle) + c*np.cos(angle), c*np.cos(angle), 0]
plt.fill(square_c_x, square_c_y, alpha=0.3, color='green', label=f'c² = {c**2:.1f}')

plt.title(f'Pythagorean Theorem: {a}² + {b}² = {c:.1f}²', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"✓ {a**2} + {b**2} = {c**2:.1f}")
```
